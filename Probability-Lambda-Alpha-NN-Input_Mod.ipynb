```python
#imports 
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

#import new_probability_readin as data
import data_read_inputmod as data
```

```python
#load input data for function
#set up training data
training_vals = data.getTrain() #indexing the number of inputs
training = training_vals[0][4]
labels = training_vals[1][4]

#set up testing data
test_data = data.getTest()
testing = test_data[0][4]
test_labels = test_data[1][4]

#increase number of x values by factors of 2 2^n +1 data points to get a center point (odd) to see difference 
#scale hidden with inputs 
```

```python
print(len(training[0]))
```

```python
#set arrays to store training and testing histories 
num_input = len(training[0])

#info = data.getInfo()
#num_input = info
#training_vals = np.zeros((num_input,))
#testing_vals = np.zeros((num_input,)) 
```

```python
#set model #Scan ratios for a fixed 129 inputs and change hidden units 
#first try varying number of inputs and then look to see if 1:1 is the best ratio
model = tf.keras.models.Sequential() #feed forward
#input layer: 
model.add(tf.keras.layers.Input(num_input))
#hidden layers: 
model.add(tf.keras.layers.Dense(num_input-1, activation = tf.nn.relu)) #128 is the best: least loss
#output layer: 
model.add(tf.keras.layers.Dense(2,))


#optimize (train) 
model.compile(optimizer = 'adam', 
      loss = 'mean_squared_error',
      metrics = [tf.keras.metrics.RootMeanSquaredError()])
model.fit(training, labels, epochs = 600)
model.evaluate(x = testing, y = test_labels)
```

```python
#saves the model data
model.save('probability_nn.model')
new_model = tf.keras.models.load_model('probability_nn.model')
```

```python
predictions = new_model.predict(testing)
print(predictions.shape)
alpha_predictions = predictions.T[0]
lambda_predictions = predictions.T[1] 
print(alpha_predictions.shape)
print(lambda_predictions.shape)
```

```python
for i in range(len(alpha_predictions)):
    if(alpha_predictions[i]-test_labels.T[0][i] >= 0.1):
        print(test_labels[i])
    elif(lambda_predictions[i]-test_labels.T[1][i] >= 0.1):
        print(test_labels[i])
```

```python
#graph alpha error

plt.scatter(test_labels.T[0], alpha_predictions) #alpha
plt.title('Predicted Alpha vs. Actual Alpha: 129I, 128H')
plt.xlabel('Actual Alpha')
plt.ylabel('Predicted Alpha')
plt.show() 
```

```python
plt.hist(alpha_predictions-test_labels.T[0])
plt.title('Distribution of Alpha Errors: 129I, 128H')
plt.ylabel('Number of Occurences')
plt.xlabel('Absolute Error')
plt.show()
```

```python
#graph lambda error
plt.scatter(test_labels.T[1], lambda_predictions) #lambda
plt.title('Predicted Lambda vs. Actual Lambda: 129I, 128H')
plt.xlabel('Actual Lambda')
plt.ylabel('Predicted Lambda')
plt.show() 
```

```python
plt.hist(lambda_predictions-test_labels.T[1])
plt.title('Distribution of Lambda Errors: 129I, 128H')
plt.ylabel('Number of Occurences')
plt.xlabel('Absolute Error')
plt.show()
```
